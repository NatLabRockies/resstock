```python
from orchestrator import PlotOrchestrator

from resstockpostproc.standard_plots.schema.plot_spec import (
    ComparisonTypes,
    PlotSpec,
    UpgradeInclusion,
    VacancyInclusion,
    VizType,
)
from resstockpostproc.standard_plots.schema.workflow_schema import QuantityGroup
```

```python
# auto reload modules
%load_ext autoreload
%autoreload 2
```

```python
def get_quantity_group(plot_orchestrator: PlotOrchestrator, quantity_group_name: str) -> QuantityGroup:
    for qg in plot_orchestrator.workflow.quantities:
        if qg.name == quantity_group_name:
            return qg
    raise ValueError(f"Quantity group {quantity_group_name} not found")
```

```python
po = PlotOrchestrator("workflow.yaml")
plot_spec = PlotSpec(
    upgrade_inclusion=UpgradeInclusion.all,
    vacancy_inclusion=VacancyInclusion.all,
    comparison_type=ComparisonTypes.absolute,
    visualization_type=VizType.box,
    group_by="in.geometry_building_type_recs",
    quantity="out.site_energy.total.energy_consumption.kwh",
    quantity_group_name="net_energy_consumption",
)

plotter = PlotOrchestrator.get_plotter(plot_spec.visualization_type)
folder_segment, name = plot_spec.get_path_and_name()
df = po.processor.prepare_data_for_plot(plot_spec)
fig = plotter.create_plot(df, plot_spec)
fig.show()
```
